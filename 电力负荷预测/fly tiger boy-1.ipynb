```python
from datetime import datetime

def get_season(date):
    if (date.month == 12 and date.day >= 21) or (date.month == 1 or date.month == 2):
        return "Winter"
    elif (date.month == 3 or date.month == 4 or date.month == 5):
        return "Spring"
    elif (date.month == 6 or date.month == 7 or date.month == 8):
        return "Summer"
    else:  # (date.month == 9 or date.month == 10 or date.month == 11):
        return "Autumn"
```
